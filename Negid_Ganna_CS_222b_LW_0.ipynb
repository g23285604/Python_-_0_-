# Таблиця 1. Завдання до виконання роботи

```python
def gcd_division(a, b):
    while b:
        a, b = b, a % b
    return a

def gcd_subtraction(a, b):
    while a != b:
        if a > b:
            a -= b
        else:
            b -= a
    return a

def task_2():
    try:
        a, b = map(int, input("Введіть два числа через пробіл: ").split())

        if a <= 0 or b <= 0:
            print("Числа повинні бути більше нуля.")
            return

        gcd_d = gcd_division(a, b)
        print(f"Найбільший спільний дільник {a} і {b} за допомогою ділення – це {gcd_d}")

        gcd_s = gcd_subtraction(a, b)
        print(f"Найбільший спільний дільник {a} і {b} за допомогою віднімання – це {gcd_s}")
    
    except ValueError:
        print("Помилка! Будь ласка, введіть два цілих числа.")

if __name__ == "__main__":
    task_2()
```
